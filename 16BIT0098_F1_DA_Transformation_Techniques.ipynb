```
fig2=plt.figure(figsize=(18,18))

img2name='drive/My Drive/Academics/Transformation Techniques/pepper.jpg'

img2=mpimg.imread(img2name)
fig2.add_subplot(131)
plt.imshow(img2, alpha=1)

limg2=list(img2)

gmi2=[]
for x in range(len(limg2)):
  temp=[]
  for y in range(len(limg2[x])):
    tempVal=255-limg2[x][y]
    tempList=[255-tempVal,255-tempVal,255-tempVal]
    temp.append(tempList)
  gmi2.append(temp)
  
fig2.add_subplot(132)
plt.imshow(gmi2, alpha=1)

half=[]
for i in range(len(limg)):
  temp=[]
  for j in range(int(len(limg[i])/2)):
    temp.append(limg[i][j])
  for k in range(int(len(limg[i])/2),len(gmi2)):
    temp.append(gmi2[i][k])
  half.append(temp)
  
fig2.add_subplot(133)
plt.imshow(half, alpha=1)

plt.show()
```

```
fig3=plt.figure(figsize=(12,12))

img3name='drive/My Drive/Academics/Transformation Techniques/lenacolor.jpg'

img3=mpimg.imread(img3name)
fig3.add_subplot(121)
plt.imshow(img3, alpha=1)

limg3=list(img3)

swapped=[]
for i in limg3:
  temp=[]
  for j in i:
    temptemp=[]
    temptemp.append(j[1])
    temptemp.append(j[2])
    temptemp.append(j[0])
    temp.append(temptemp)
  swapped.append(temp)
  
fig3.add_subplot(122)
plt.imshow(swapped, alpha=1)

plt.show()
```
